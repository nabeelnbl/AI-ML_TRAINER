```python
#combine () - combine 2 DF objects use a specified function to decide which column to keep
df1=pd.DataFrame([[10,2],[3,4]])
df2=pd.DataFrame([[5,6],[7,8]])
def myfunc(a,b):
    if (a.sum()>b.sum()):
        return a
    else:
        return b
print(df1.combine(df2,myfunc))
```
